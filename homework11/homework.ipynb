```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

## Дана функция: f(x) = x^2 - 6*abs(x) + 8

```python
x = Symbol("x", real=True)
f = x**2 - 6*abs(x) + 8
f
```

1. Определить корни

```python
solve(f)
```

2. Найти интервалы, на которых функция возрастает

```python
d = diff(f)
d
```

```python
d_root = solve(d)
d_root = [-3, 0, *d_root]
d_root
```

```python
plot(d)
```

```python
solve(d > 0)
```

3. Найти интервалы, на которых функция убывает

```python
solve(d < 0)
```

4. Построить график

```python
p = plot(f)
p
```

5. Вычислить вершину

```python
s_result = [(i, f.subs(x, i)) for i in d_root]
s_result
```

6. Определить промежутки, на котором f > 0

```python
solve(f > 0)
```

7. Определить промежутки, на котором f < 0

```python
solve(f < 0)
```
